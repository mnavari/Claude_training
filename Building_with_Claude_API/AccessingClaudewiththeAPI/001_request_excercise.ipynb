```python
# load environment variables from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Createan API client
from anthropic import Anthropic
client = Anthropic()    
model = "claude-sonnet-4-0" 
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user", "content": text}
    messages.append(user_message)   

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content": text}
    messages.append(assistant_message) 

# Make a request to the API
def chat(messages):  
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        #messages=[
        #    {"role": "user", "content": "What is quantum computing? Answer in one sentence." }]
        messages=messages
    )
    return message.content[0].text

   
```

```python
# Make a string list of messages
messages = []
# add in the inital user question  
add_user_message(messages, "What is quantum computing? Answer in one sentence.")
# pass in the messages list to the chat function
answer = chat(messages)
# take the answer and add it to the messages list as an assistant message
add_assistant_message(messages, answer)
# Add in the users follow up question
add_user_message(messages, "What are the applications of quantum computing?")
# pass in the updated messages list to the chat function
answer = chat(messages)
print(answer)   
```

```python
# make an initial list of messages
messages = []

# Use a 'While True' loop to run the chatbot forever
while True: 
    # Get user input and add it to the messages list
    user_input = input("User: ")
    print(f"User: {user_input}")

    # Add the user's message to the messages list
    add_user_message(messages, user_input)

    # Call Claude with the 'Chat' function
    answer = chat(messages)

    # Add generated response to the messages list
    add_assistant_message(messages, answer)

    # Print the generate text 
    print("---")
    print(f"Claude: {answer}") 
    print("---") 
```
