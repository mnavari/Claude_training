```python
# load environment variables from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Createan API client
from anthropic import Anthropic
client = Anthropic()    
model = "claude-sonnet-4-0" 
```

```python
# adding temperature to the request to increase creativity
def add_user_message(messages, text):
    user_message = {"role":"user", "content": text}
    messages.append(user_message)   

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content": text}
    messages.append(assistant_message) 

# Make a request to the API
def chat(messages, system = None, stop_sequences=None):
    parameters = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }   
    if system: 
        parameters["system"] = system

    message = client.messages.create(**parameters)
    return message.content[0].text

```

```python
messages = []

add_user_message(messages, "write a one sentence description of a fake database")

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)   
for event in stream:
    print(event)    
```
