```python
# load environment variables from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Createan API client
from anthropic import Anthropic
client = Anthropic()    
model = "claude-sonnet-4-0" 
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user", "content": text}
    messages.append(user_message)   

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content": text}
    messages.append(assistant_message) 

# Make a request to the API
def chat(messages, system = None):
    parameters = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }   
    if system: 
        parameters["system"] = system

    message = client.messages.create(**parameters)
    return message.content[0].text
```

```python
messages = []

system = """"You are a patient math tuter. 
Do not directly answer a student's questions.
Guide the them to a solution step by step
"""
add_user_message(messages, "How do I solve 5X + 3=2 for X?")
answer = chat(messages, system = system)

answer
```

```python
# adding temperature to the request to increase creativity
def add_user_message(messages, text):
    user_message = {"role":"user", "content": text}
    messages.append(user_message)   

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content": text}
    messages.append(assistant_message) 

# Make a request to the API
def chat(messages, system = None, temperature = 1.0):
    parameters = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }   
    if system: 
        parameters["system"] = system

    message = client.messages.create(**parameters)
    return message.content[0].text

```
